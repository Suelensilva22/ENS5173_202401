```python
import numpy as np #importando a biblioteca NumPy com o nome np 

# Criando um vetor de dados aleatórios
data = np.random.rand(1000,2) #cria uma matriz NumPy com 1000 linhas e 2 colunas chamada "data", onde cada elemento é um número aleatório entre 0 e 1

# Verificando o tamanho do vetor
print(data.shape) #escreve a forma da matriz "data"
print(np.size(data)) #escreve o número total de elementos na matriz "data"

# Print de uma amostra dos dados
print(data[0:10]) #acessando da linha 0 até a Linha 9 (10 primeiras linhas da matriz)
print(data[0:10].shape) #retorna a forma (shape) de uma fatia da matriz "data"
```

```python
# Usando Monte-Carlo
dataMax = 1000 #Definindo a variável "dataMax"
dataMin = 10 #Definindo a variável "dataMin"
data = (dataMax - dataMin)*data + dataMin #escala os valores da matriz "data" para o intervalo entre "dataMin" e "dataMax"
print(data[0:10]) #acessando da linha 0 até a Linha 9 (10 primeiras linhas da matriz) depois da transformação
```

```python
import matplotlib.pyplot as plt #importa a biblioteca Matplotlib para criação de gráficos em Python

fig, ax = plt.subplots(3) #cria uma figura com 3 subplots (é uma função em Matplotlib que é usada para criar várias áreas de plotagem em uma única figura) organizados em uma coluna
ax[0].plot(data) #plota todos os dados
ax[1].plot(data[:,0]) #plota apenas a primeira coluna
ax[2].plot(data[:,1]) #plta apenas a segunda coluna
print(type(data)) #escreve o tipo da variável 'data'
```

```python
# Criando um dado 2D 
data3d = np.sin(np.random.rand(1000,1000)) #cria uma matriz NumPy 1000x1000 e calcula o seno para cada elemento dessa matriz, armazenando o resultado na variável "data3d"
print(data3d) #escrever a matriz "data3d"
```

```python
# Plotando um dado 3D
fig2, ax2 = plt.subplots() # cria um novo subplot (um espaço para plotagem) e armazena a figura e o eixo correspondente nas variáveis fig2 e ax2
ax2.pcolor(data3d) #criar um gráfico de mapa de cores 2D a partir dos dados contidos na matriz "data3d"
```

```python
# Criando vertor temporal
dtime = np.arange(np.datetime64("2022-01-01"), np.datetime64("2023-01-01"), np.timedelta64(1, "h")) #cria uma variedade de datas e horas utilizando o NumPy, começa em "2022-01-01" e termina em "2023-01-01" (intervalos de uma hora)
print(dtime[1:10]) #mostra as datas e horas a partir da segunda (índice 1) até a décima (índice 9) na variável "dtime"
print(dtime.shape) #mostra quantos elementos tem na variável "dtime"
print(type(dtime)) #escrever o tipo de variável é "dtempo".
```

## Pandas

```python
import pandas as pd #importando a biblioteca Pandas com o nome pd 

df = pd.DataFrame() #criando um dataframe (quadro de dados) utilizando a biblioteca pandas chamado "df"
```

```python
df['datetime'] = dtime #monstrando um dataframe onde a variável dtime é atribuida à coluna "datetime" do "df"
```

```python
df['random'] = np.random.rand(df.shape[0]) #criando uma nova coluna chamada "random" no "df" e preenche essa coluna com valores aleatórios gerados pela função "np.random.rand()" com base no número de linhas do DataFrame
```

```python
df['random'][1:10] #acessa as linhas de índice 1 a 9 na coluna "random" do "df"
```

```python
# Indexação
df.iloc[:,0] #seleciona todas as linhas da primeira coluna do "df" utilizando indexação baseada em posição 
```

```python
# Colocando o datetime como index
df = df.set_index(df['datetime']) #redefine o índice do "df" para a coluna "datetime", ou seja, os valores dessa coluna se tornam os índices do DataFrame.
df #o DataFrame "df" será exibido com a coluna "datetime" como seu índice
```

```python
# Criando colunas com ano, mês e dia
df['year'] = df.index.year #criando a coluna ano
df['month'] = df.index.month #criando a coluna mês
df['day'] = df.index.day #criando a coluna dia

#os valores dessas colunas são extraídos do índice do DataFrame, que foi definido anteriormente como a coluna "datetime"
```

```python
tt = np.arange(0,df.shape[0],1) #cria uma sequência tt de números inteiros de 0 ao número de linhas no "df"
print(tt.shape) #mostra quantos elementos tem na variável "tt"
df['sazonal'] = (np.cos(tt/(360*4))) #cria uma coluna sazonal no "df" utilizando a função cosseno (np.cos) aplicada à sequência tt, gerando uma componente sazonal
df['timeSeries'] = 10*df['sazonal']+df['random'] #cria uma coluna "timeSeries" no "df", que é a combinação da componente sazonal multiplicada por 10 e a coluna random
#df['sazonal'].plot()
df['timeSeries'].plot(color='red') #plota a coluna "timeSeries" em um gráfico de linha com a cor vermelha.
```
