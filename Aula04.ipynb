```python
#Criando uma lista em Python

list = [0, 'eu amo polAtm', 2.5]
print(list)
print('Toda lista:'+str(list)) #pra somar texto com outros tipos de variáveis
print('variável:'+list[1]) #aqui deu pra somar sem str porque a posição 2 [1] da lista é str
print(list[0]+1) #aqui deu pra somar sem str porque a posição 1 [0] da lista é número, indexação no python inicia em 0
```

```python
#Criando uma matriz numpy

import numpy as np #importando o módulo numpy como "np"

a = np.zeros((5,5)) #criando uma matriz de zero utilizando o np.zeros(( ))
print(a) #mostrar a matriz "a"
print('indexação numpy')
print(a[0,0]) #mostrar uma posição específica da matriz
```

```python
#Criando números aleatórios e mostrando a dimensão da matriz

b = np.random.rand(5,5) #matriz de números aleatórios "b" com np.random.rand( )
print(b) #mostrar a matriz "b"
print('Dimensão da matriz') #mostrar a frase dentro do (' ')
print(b.shape) #mostrar o shape da matriz, ou seja a dimensão da mesma
```

```python
#Método Monte Carlo: vamos usar para criar uma matriz dentro de uma faixa de valores
#Vmin+(Vmax-Vmin)*matriz de numeros aleatorios
```

```python
#Criando matriz de valores aleatórios entre uma faixa estabelecida

Vmin = 1 #valor mínimo
Vmax = 100 #valor máximo

c = Vmin+(Vmax-Vmin)*b
print(c)
```

```python
# Plotar matriz

import matplotlib.pyplot as plt #impportando o pacote pyplot do módulo matplotlib chamando de plt

plt.plot(c) #plotar o gráfico da matriz
```

```python
plt.plot(c.transpose()) #plotar o gráfico da matriz transposta pq o "c" é uma matriz numpy
```

```python
# Plotando só a primeira linha
plt.plot(c[0,:]) #plotando a primeira linha e todas as colunas 
```

```python
#plotando com valores do eixo x
x=range(np.size(c[0,:]))
print(x)

plt.plot(c[0,:]) #plotando a primeira linha e todas as colunas 
```

```python
#posso transformar pra uma matriz numpy
x=np.array(range(np.size(c[0,:])))
print(x)

plt.plot(c[0,:]) #plotando a primeira linha e todas as colunas 
```

```python
#deixar o eixo x com números inteiros
x=np.array(range(np.size(c[0,:])))
print(x)

plt.plot(c[0,:]) #plotando a primeira linha e todas as colunas 
plt.title('Gráfico com Valores Inteiros no Eixo X')
plt.xticks(x) #Definindo os valores do eixo x
```

```python
#Criando um gráfico de barras e exportando

fig, ax = plt.subplots(3,1) #3 linhas e 1 coluna
ax[0].bar(x,c[:,0]) #dentro do eixo 0, coloque (x,c[:,0]) 
ax[1].bar(x,c[:,1]) #dentro do eixo 1, coloque (x,c[:,1])
ax[2].bar(x,c[:,2]) #dentro do eixo 2, coloque (x,c[:,2])

#O "bar" indica que é um gráfico de barras e ax é um objeto plt

#exporta a figura para a pasta ENS
#fig.savefig('figure01.png') 

#exporta a figura para a pasta figures no ENS
import os
rootFolder = os.getcwd()
print(rootFolder)
figFolder = rootFolder+r'\\figures\\'
print(figFolder)
fig.savefig(figFolder+'figure01.png') 
```
